### Multiple Season

```python
#creating a string to work with multiple string
for season in range(20, 26):
    print(str(season) + str(season + 1))
```

```python
#Creating a list og leagues
leagues = ["E0", "E2", "E3"]
frames = []

#Looping through leagues, read multiple csv and append it to the list
for league in leagues:
    for season in range(20, 26):
        df = pd.read_csv(root + str(season) + str(season + 1) + "/" + league + ".csv")
        df.insert(1, 'season', season)
        frames.append(df)
```

## Organize all the data in a Dictionary

```python
#Create a country with the original name of leagues as key
dict_countries = {"Spanish La Liga" : "SP1", "Spanish Segunda Division" : "SP2",
                  "English Premier League" : "E0", "English League 1" : "E2", "English League 2" : "E3"}
```

```python
#Get a Dictionary Element
dict_countries['Spanish La Liga']
```

```python
#How iteration in dictionaries work
for league in dict_countries:
    print(league)
```

```python
#How iteration in dictionaries work
for league in dict_countries:
    print(dict_countries[league])
```

```python
dict_historical_data = {}

#Creating a list og leagues
#leagues = ["E0", "E2", "E3"]

#Looping through key elements
for league in dict_countries:
    frames = [] #Everytime this loops starts list initializes
    for season in range(20, 26):
        df = pd.read_csv(root + str(season) + str(season + 1) + "/" + dict_countries[league] + ".csv")
        df.insert(1, 'season', season)
        frames.append(df)
    df_concat = pd.concat(frames) # Concatinates all the Dataframes of a league from diff. seasons
    dict_historical_data[league] = df_concat #Adds all the data into a Dictionary
```
