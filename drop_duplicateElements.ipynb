```python
import pandas as pd
df_laptops = pd.read_csv('../laptop_price.csv', encoding='latin1')
df_laptops.head()
```

```python
# Sorting according to Company and Inches (Ascending)
df_laptops = df_laptops.sort_values(['Company', 'Inches'])
```

```python
#dropping duplicates, Keeping Smallest Screen Size
df_laptops.drop_duplicates(['Company'], keep='first', ignore_index=True)[['Company', 'Inches']]
```

```python
#dropping duplicates, Keeping Biggest Screen Size
df_laptops.drop_duplicates(['Company'], keep='last', ignore_index=True)[['Company', 'Inches']]
```
